```python

sns.scatterplot(
    x=or_data['input1'],
    y=or_data['input2'],
    hue=or_data['output'],
    palette='winter',
    s=100
)
```

```python
from sklearn.linear_model import Perceptron
```

```python
clf1=Perceptron()
clf2=Perceptron()
clf3=Perceptron()
```

```python
clf1.fit(and_data.iloc[:,0:2],and_data.iloc[:,-1].values)
clf2.fit(or_data.iloc[:,0:2],or_data.iloc[:,-1].values)
clf3.fit(xor_data.iloc[:,0:2],xor_data.iloc[:,-1].values)
```

```python
clf1.coef_
```

```python
clf1.intercept_
```

```python
x=np.linspace(-1,1,5)
y=-(clf1.coef_[0][0]*x+clf1.intercept_)/clf1.coef_[0][1]#y=mx+c
```

```python
plt.plot(x,y)
sns.scatterplot(
    x=and_data['input1'],
    y=and_data['input2'],
    hue=and_data['output'],
    s=200
)
```
