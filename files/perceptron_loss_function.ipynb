```python
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt 
```

```python
X,y=make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=15)
```

```python
X.shape
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=X[:,0],y=X[:,1],c=y,cmap='bwr')
```

```python
def perceptron(X,y):
    w1=w2=b=1
    lr=0.1
    for j in range(100):
        for i in range(X.shape[0]):
            z=w1*X[i][0]+w2*X[i][1]+b
            if z*y[i]<0:#only update when the points are misclassified
                w1 = w1 + lr*y[i]*X[i][0]#apply 
                w2 = w2 + lr*y[i]*X[i][1]
                b = b + lr*y[i]
    return w1,w2,b

```

```python
w1,w2,b=perceptron(X,y)
print(w1,w2,b)
```

```python
m=-(w1/w2)
coef_=-(b/w2)
```

```python
print("slope:", m,'\n' "intercept:", coef_) 
```

```python
x_input=np.linspace(-3,3,100)
y_input=m*x_input+coef_#y=mx+c
```

```python
plt.figure(figsize=(12,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.plot(x_input,y_input,color='black')
```
