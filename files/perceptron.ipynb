```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=10)
```

```python
X.shape
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter')
```

```python
X=np.insert(X,0,1,axis=1)
```

```python
X
```

```python
def perceptron(X,y):
    X=np.insert(X,0,1,axis=1)# we inser a column of 1 at the 0 index 
    w=np.zeros(X.shape[1])# # intialize the weights to 0 (the number of columns )
    lr=0.1#learing rate
    for i in range(1000):
        j=np.random.randint(0,X.shape[0])# we select a random index from the number of rows
        y_hat=step(np.dot(X[j],w))
        w+=lr*(y[j]-y_hat)*X[j]
    return w

    
```

```python
a=perceptron(X,y)
```

```python
a

#w = [w0, w1, w2]
'''w[0] = 0.5 → bias (intercept)
w[1] = 0.678... → weight for feature 1
w[2] = -0.023... → weight for feature 2'''

```

```python
def step(Z):
    if Z>=0:
        return 1
    else:
        return 0
```

```python
a
```

```python
m=-(a[1]/a[2])
coef_=-(a[0]/a[2])
print("slope:", m,'\n' "intercept:", coef_)
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + coef_
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
